```python
import pandas as pd
```

```python
positive_housekeeping_text = pd.read_csv('positive_Kate_housekeeping.csv')
positive_housekeeping_text.head()
```

```python
positive_warehouse_text = pd.read_csv('positive_masha.csv')
positive_warehouse_text.head()
```

```python
# positive_warehouse_text['complain_real'] = positive_warehouse_text.apply(lambda x: x, axis=0)
# positive_warehouse_text.head()
cols = list(positive_warehouse_text.columns)
positive_warehouse_text['combined'] = positive_warehouse_text[cols].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
```

```python
positive_warehouse_text = positive_warehouse_text[['class', 'combined']]
positive_warehouse_text = positive_warehouse_text.rename(columns={'combined':'complain'})
positive_warehouse_text.head()
```

```python
positive_compl_ashley = pd.read_csv('Complaints  - Ashley.csv')
positive_compl_ashley = positive_compl_ashley.rename(columns={'Class':'class', 'Complaint':'complain'})
positive_compl_ashley.head()
```

```python
positive_compl_albin = pd.read_csv('albin_positives.csv').rename(columns={'Class':'class', 'Complain':'complain'})
positive_compl_albin.head()
```

```python
positive_combined = pd.concat([positive_housekeeping_text, positive_warehouse_text, positive_compl_ashley, positive_compl_albin], axis=0)
positive_combined.head()
```

```python
positive_combined.to_csv('positive_combined.csv', encoding='utf-8', index=False)
```
